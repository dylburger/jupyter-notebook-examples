```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.1)
```

```python
y = np.sin(x)
```

```python
def apply_color(x):
    return 'red' if x > 0 else 'blue'
    
colors = [apply_color(point) for point in y]

# plt.scatter can take a vector of colors, which are strings
# tied to two colors of interest here ('red' or 'blue')
plt.scatter(x, y, c=colors)

plt.show()
```
